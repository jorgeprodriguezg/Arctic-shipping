```python
import odp.geospatial as odp
import pandas as pd
```

```python
gd=odp.GridData()
```

```python
ds=gd.open_dataset('Vessel Traffic and CO2 Emissions')
```

```python
utimes = list(ds.time.values)
print(utimes,len(utimes))
```

```python
ds_arctic = ds.sel(lat=slice(66.4,90.0),time=slice('2019-01-01','2022-12-31')
                ).hours_total.compute()
```

```python
totalh = []
for t in utimes:
    ds2 = ds_arctic.sel(time=t)
    totalh.append(ds2.sum().values)
```

```python
import pylab as plt
plt.style.use('classic')
```

```python
plt.plot(utimes,totalh)
plt.xticks(rotation=90)
```

```python
for t,val in zip(utimes,totalh):
    print(t,val)
```

```python
del ds_arctic,ds2
```

```python
ds_arctic = ds.sel(lat=slice(66.4,90.0),time=slice('2019-01-01','2022-12-31')
                ).hours_cargo.compute()
```

```python
totalc = []
for t in utimes:
    ds2 = ds_arctic.sel(time=t)
    totalc.append(ds2.sum().values)
```

```python
for t,val in zip(utimes,totalh):
    print(t,val)
```

```python
del ds_arctic,ds2
ds_arctic = ds.sel(lat=slice(66.4,90.0),time=slice('2019-01-01','2022-12-31')
                ).hours_fishing.compute()
```

```python
totalf = []
for t in utimes:
    ds2 = ds_arctic.sel(time=t)
    totalf.append(ds2.sum().values)
```

```python
for t,val in zip(utimes,totalf):
    print(t,val)
```

```python
del ds_arctic,ds2
ds_arctic = ds.sel(lat=slice(66.4,90.0),time=slice('2019-01-01','2022-12-31')
                ).hours_passenger.compute()
```

```python
totalp = []
for t in utimes:
    ds2 = ds_arctic.sel(time=t)
    totalp.append(ds2.sum().values)
```

```python
for t,val in zip(utimes,totalp):
    print(t,val)
```

```python
del ds_arctic,ds2
ds_arctic = ds.sel(lat=slice(66.4,90.0),time=slice('2019-01-01','2022-12-31')
                ).hours_tanker.compute()
```

```python
totalt = []
for t in utimes:
    ds2 = ds_arctic.sel(time=t)
    totalt.append(ds2.sum().values)
```

```python
for t,val in zip(utimes,totalt):
    print(t,val)
```

```python
cf = (0.00392156862745098, 0.45098039215686275, 0.6980392156862745)
cp = (0.792156862745098, 0.5686274509803921, 0.3803921568627451)
cc = (0.8, 0.47058823529411764, 0.7372549019607844)
ct = (0.9254901960784314, 0.8823529411764706, 0.2)
```

```python
fig = plt.figure(figsize=(12,15))
ax = plt.gca()
ax.set_position([0.,0.6,1,0.5])
plt.plot(utimes,[e1/24e4 for e1 in totalf],
         c=cf,
         lw=3,label='Fishing')
plt.plot(utimes,[e1/24e4 for e1 in totalp],c=cp,lw=3,label='Passenger')
plt.plot(utimes,[e1/24e4 for e1 in totalc],c=cc,lw=3,label='Cargo')
plt.plot(utimes,[e1/24e4 for e1 in totalt],c=ct,lw=3,label='Tanker')

plt.xticks(rotation=90,size=20,color='#222222')
plt.yticks(size=20,color='#222222')
plt.yscale('log')
plt.ylabel(r'Shipping traffic ($\times 10^4$ days)',size=25,color='#222222')
l = plt.legend(bbox_to_anchor=(1.25,0.9),fontsize=20)
for text in l.get_texts():
    text.set_color("#222222")
ax2 = fig.add_axes([0.,0,1,0.5])

plt.fill_between(utimes,[e1/24e4 for e1 in totalh],
                 y2=0,zorder=1,color='#999999',label='Other',ec='k')
plt.fill_between(utimes,
                 [e1/24e4+e2/24e4+e3/24e4 for e1,e2,e3 in zip(totalc,totalt,totalp)],
                 [e1/24e4+e2/24e4+e3/24e4+e4/24e4 for e1,e2,e3,e4 in zip(totalc,totalt,totalp,totalf)],
                 color=cf,
                 zorder=2,ec='k',label='Fishing')
plt.fill_between(utimes,[e1/24e4+e2/24e4 for e1,e2 in zip(totalc,totalt)],
                 [e1/24e4+e2/24e4+e3/24e4 for e1,e2,e3 in zip(totalc,totalt,totalp)],
                 color=cp,zorder=2,ec='k',label='Passenger')
plt.fill_between(utimes,[e1/24e4 for e1 in totalt],
                 [e1/24e4+e2/24e4 for e1,e2 in zip(totalc,totalt)],
                 color=cc,zorder=3,ec='k',label='Cargo')
plt.fill_between(utimes,0,[e1/24e4 for e1 in totalt],
                 color=ct,zorder=2,ec='k',label='Tanker')



plt.xticks(rotation=90,size=20,color='#222222')
plt.yticks(size=20,color='#222222')
plt.ylabel(r'Shipping traffic ($\times 10^4$ days)',size=25,color='#222222')
l1 = plt.legend(bbox_to_anchor=(1.25,0.9),fontsize=20)
for text in l1.get_texts():
    text.set_color("#222222")
fig.savefig('time_ev_bycategory.png',dpi=300,bbox_inches='tight')
fig.savefig('time_ev_bycategory.pdf',dpi=300,bbox_inches='tight',fmt='pdf')
```
