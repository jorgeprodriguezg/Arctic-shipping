```python
import pandas as pd
import pylab as plt
import numpy as np
plt.style.use('classic')
import matplotlib.colors as colors
import matplotlib.colorbar as chi
import math
```

```python
dfnew = pd.read_csv('./data/routing_tev_Ntimearea_bylon_025.csv')
```

```python
dfold = pd.read_csv('./data/routing_tev_N_bylon_025old.csv')
#This data comes from 
#Eguíluz, V. M., Fernández-Gracia, J., Irigoien, X., & Duarte, C. M. (2016). 
#A quantitative assessment of Arctic shipping in 2010–2014. Scientific reports, 6(1), 30682.
```

```python
dfold.head()
```

```python
dfnew.head()
```

```python
ilon1 = 4*(-90+180)
ilon2 = 4*(150+180)
```

```python
dl = 6371.*0.25*np.pi/180.0
```

```python
wseries = []
samp = dfold[dfold['ilon'] == ilon1]
y = 2010
for m in range(7,13):
    if m < 10:
        mo = f'0{m}'
    else:
        mo = str(m)
    wseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
for y in range(2011,2015):
    for m in range(1,13):
        if m < 10:
            mo = f'0{m}'
        else:
            mo = str(m)
        wseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
y = 2015
for m in range(1,6):
    if m < 10:
        mo = f'0{m}'
    else:
        mo = str(m)
    wseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
```

```python
plt.plot(wseries)
```

```python
eseries = []
samp = dfold[dfold['ilon'] == ilon2]
y = 2010
for m in range(7,13):
    if m < 10:
        mo = f'0{m}'
    else:
        mo = str(m)
    eseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
for y in range(2011,2015):
    for m in range(1,13):
        if m < 10:
            mo = f'0{m}'
        else:
            mo = str(m)
        eseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
y = 2015
for m in range(1,6):
    if m < 10:
        mo = f'0{m}'
    else:
        mo = str(m)
    eseries.append(samp[f'{y}_{mo}_N'].iloc[0]*dl)
```

```python
plt.plot(wseries)
plt.plot(eseries)
```

```python
fdates = list(dfnew.columns)
fdates.remove('ilon')
fdates.sort()
fdates2 = []
for elem in fdates:
    if elem[-1] == 'N':
        fdates2.append(elem)
fdates3 = []
for elem in fdates2:
    fdates3.append(elem[:-2])
print(fdates3,len(fdates3))
```

```python
wseriesn = []
samp = dfnew[dfnew['ilon'] == ilon1]
for d in fdates2:
    wseriesn.append(samp[d].iloc[0]*dl)
```

```python
eseriesn = []
samp = dfnew[dfnew['ilon'] == ilon2]
for d in fdates2:
    eseriesn.append(samp[d].iloc[0]*dl)
```

```python
ndates = []
for y in range(2020,2022):
    for m in range(1,13):
        ndates.append(pd.to_datetime(f'{y}-{m}-01'))
y = 2022
for m in range(1,5):
    ndates.append(pd.to_datetime(f'{y}-{m}-01'))
```

```python
odates = []
y = 2010
for m in range(7,13):
    odates.append(pd.to_datetime(f'{y}-{m}-01'))

for y in range(2011,2015):
    for m in range(1,13):
        odates.append(pd.to_datetime(f'{y}-{m}-01'))
y = 2015
for m in range(1,6):
    odates.append(pd.to_datetime(f'{y}-{m}-01'))
```

```python
plt.plot(odates,eseries)
plt.plot(odates,wseries)
plt.plot(ndates,eseriesn)
plt.plot(ndates,wseriesn)
```

```python
f,(ax,ax2) = plt.subplots(1,2,sharey=True, facecolor='w')
ax.set_position([0,0,1,1])
ax2.set_position([1.05,0,28./59,1])
plt.sca(ax)
plt.plot(odates,eseries,c='#4575b4',lw=2)
plt.plot(odates,wseries,c='#d73027',lw=2)
plt.plot(ndates,eseriesn,c='#4575b4',lw=2)
plt.plot(ndates,wseriesn,c='#d73027',lw=2)
plt.xticks(fontsize=18,color='#444444')
plt.yticks(fontsize=18,color='#444444')
plt.ylabel(r'Route width (km)',size=25,color='#444444')
plt.sca(ax2)
plt.plot(odates,eseries,c='#4575b4',lw=2)
plt.plot(odates,wseries,c='#d73027',lw=2)
plt.plot(ndates,eseriesn,c='#4575b4',lw=2,label='Northeastern Route (150$^{\circ}$)')
plt.plot(ndates,wseriesn,c='#d73027',lw=2,label=r'Northwest Passage (-90$^{\circ}$)')

l = plt.legend(loc=0,prop={'size':16})
for elem in l.get_texts():
    elem.set_color('#444444')
ax.set_xlim(pd.to_datetime('2010-06-01'), pd.to_datetime('2015-06-01'))
ax2.set_xlim(pd.to_datetime('2020-01-01'), pd.to_datetime('2022-04-01'))


# hide the spines between ax and ax2
ax.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax.spines['left'].set_color('#444444')
ax.spines['top'].set_color('#444444')
ax.spines['bottom'].set_color('#444444')

ax2.spines['right'].set_color('#444444')
ax2.spines['top'].set_color('#444444')
ax2.spines['bottom'].set_color('#444444')
ax.yaxis.tick_left()
ax2.tick_params(labelright='off',labelleft='off')
ax2.yaxis.tick_right()
plt.yticks(fontsize=18,color='#444444')

d = .015 # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='#444444', clip_on=False)
ax.plot((1-d,1+d), (-d,+d), **kwargs)
ax.plot((1-d,1+d),((1-d),(1+d)), **kwargs)

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d/28*59,+d/28*59), (1-d,1+d), **kwargs)
ax2.plot((-d/28*59,+d/28*59), (-d,+d), **kwargs)
# ax2.set_xticks([pd.to_datetime('2020-01-01'),pd.to_datetime('2021-01-01'),
#                pd.to_datetime('2022-01-01')],[2020,2021,2022])
plt.sca(ax2)
plt.xticks([pd.to_datetime('2020-01-01'),pd.to_datetime('2021-01-01'),
           pd.to_datetime('2022-01-01')],[2020,2021,2022],fontsize=18,color='#444444')
f.savefig('fig7.png',bbox_inches='tight',dpi=300)
f.savefig('fig7.pdf',bbox_inches='tight',dpi=300);
```
