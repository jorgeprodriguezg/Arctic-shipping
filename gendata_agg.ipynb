```python
import geopandas as gp
import numpy as np
```

```python
clat = []
for lat in ds_arctic.lat.values:
    clat.append(int(10*(90+lat)))
clon = []
for lon in ds_arctic.lon.values:
    clon.append(int(10*(180+lon)))
print(len(clat),len(clon))
```

```python
Nlat = len(clat)
Nlon = len(clon)
```

```python
cells = []
hours = []
for i in range(Nlat):
    for j in range(Nlon):
        val = ds_arctic[i,j].values
        if val != 0.0 and val != np.nan:
            # print(val)
            c = clon[j]+10*360*clat[i]
            if c in cells:
                pt = cells.index(c)
                hours[pt]  = float(hours[pt])+float(val)
            else:
                cells.append(c)
                hours.append(float(val))
```

```python
print(len(cells))
print(sum(hours),ds_arctic.sum())
```

```python
print(hours[-10:])
```

```python
from shapely.geometry import Polygon
```

```python
lats = []
lons = []
pols = []
for c in cells:
    cy = int(c/(10*360))
    cx = c-10*360*cy
    lat = -90.0+0.1*cy
    lon = -180.0+0.1*cx
    lats.append(lat)
    lons.append(lon)
    pols.append(Polygon([(lon,lat),(lon+0.1,lat),(lon+0.1,lat+0.1),(lon,lat+0.1)]))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'lat':lats,'lon':lons,'hours_total':hours,'geometry':pols})
df.head()
```
